```python
import pandas as pd
import seaborn as sns
```

```python
# 데이터 셋 로드
df = sns.load_dataset("iris")
df.head()
```

```python
# 세 개의 컬럼을 기준으로 가져오기

df[['sepal_length','sepal_width','species']]
```

```python
columns = ['sepal_length','sepal_width','species']
df[columns].head()
```

```python
# 특정 컬럼 하나만 가져올 때

df['sepal_length']
```

```python
df.sepal_length
```

### 정규표현식을 사용해서 컬럼 추출하기

```python
# _가 들어있는 컬럼 가져오기

df.filter(regex="_")
```

```python
# length 로 끝나는 컬럼 가져오기

df.filter(regex="length$")
```

```python
# Sepal 로 시작하는 컬럼 가져오기

df.filter(regex="^sepal")
```

```python
# se로 시작하는 컬럼

df.filter(regex="^se")
```

```python
# 모든 행에 대해서 컬럼에 속하는 정보 가져오기

df.loc[:, 'sepal_width':'petal_width']
```

```python
#  2~5 행에 대해서 컬럼에 속하는 정보 가져오기
# loc : 뒤에 있는 index 포함

df.loc[2:5, 'sepal_width':'petal_width']
```

```python
# iloc : 뒤에 있는 인덱스는 포함하지 않음

df.iloc[:,[1,2,4]]
```

```python
# 앞에 있는 부분이 행, 뒤 부분이 열을 의미
# 행을 가져올 기준, 뒤는 가져올 열

df.loc[df['sepal_length'] > 3, ['sepal_length','sepal_width']]
```
