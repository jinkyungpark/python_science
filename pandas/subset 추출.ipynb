### nlargest, nsmallest

```python
df.nlargest(3,"sepal_length")
```

```python
df.nsmallest(2,"petal_length")
```

### loc, iloc

```python
# 모든 행에 대해서 컬럼에 속하는 정보 가져오기

df.loc[:, 'sepal_width':'petal_width']
```

```python
#  2~5 행에 대해서 컬럼에 속하는 정보 가져오기
# loc : 뒤에 있는 index 포함

df.loc[2:5, 'sepal_width':'petal_width']
```

```python
# iloc : 뒤에 있는 인덱스는 포함하지 않음

df.iloc[:,[1,2,4]]
```

```python
# 앞에 있는 부분이 행, 뒤 부분이 열을 의미
# 행을 가져올 기준, 뒤는 가져올 열

df.loc[df['sepal_length'] > 3, ['sepal_length','sepal_width']]
```

## 추가 개념

### 기본 추출

```python
friend_dict_list = [
    {
        "name": "John",
        "age": 25,
        "job": "student",
    },
    {
        "name": "Nate",
        "age": 30,
        "job": "teacher",
    },
    {
        "name": "Jenny",
        "age": 30,
        "job": "developer",
    },
]

friend_df = pd.DataFrame(friend_dict_list)
friend_df
```

```python
friend_df.filter(items=['name','job'])
```

```python
friend_df.filter(items=['name'])
```

```python
# 컬럼명에 'a' 가 들어간 열 추출하기

friend_df.filter(like="a", axis=1)
```

### 정규표현식을 사용해서 컬럼 추출하기

```python
# a가 들어있는 컬럼 가져오기

friend_df.filter(regex="a")
```

```python
# _가 들어있는 컬럼 가져오기

df.filter(regex="_")
```

```python
# length 로 끝나는 컬럼 가져오기

df.filter(regex="length$")
```

```python
# Sepal 로 시작하는 컬럼 가져오기

df.filter(regex="^sepal")
```

```python
# se로 시작하는 컬럼

df.filter(regex="^se")
```
