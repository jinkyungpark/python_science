### boxplot : 자료에서 얻어낸 최댓값, 최솟값, 상위 1/4, 2/4(중앙), 3/4 위치 값 이용

```python
import matplotlib.pyplot as plt
```

```python
import random
result = []
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
# 서울 최고 기온 데이터를 상자 그림으로 그리기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)


temp = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        temp.append(float(row[4]))

plt.boxplot(temp)
plt.show()
```

```python
# 1월과 8월의 기온데이터

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

temp_08 = []
temp_01 = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        month = row[0].split("-")[1]
        if month == "01": 
            temp_01.append(float(row[4]))
        if month == "08":
            temp_08.append(float(row[4]))

plt.boxplot(temp_01)
plt.boxplot(temp_08)

plt.show()
```

```python
plt.boxplot([temp_01,temp_08])
plt.show()
```

```python
# 최고 기온 데이터를 월별로 구별하여 표현하기

# 데이터를 월별로 분류해 저장 후 상자그램으로 시각화 하기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

# 월별 데이터를 담을 리스트 생성
temp = [[],[],[],[],[],[],[],[],[],[],[],[]]

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        month = row[0].split("-")[1]
        temp[int(month)-1].append(float(row[4]))  # temp[0].append() : 1월 데이터

        
# 기본으로 그리니 잘 안 보여서        
plt.figure(figsize=(20,10))   
# 그래프 스타일
plt.style.use("ggplot")

plt.boxplot(temp)
plt.show()
```

### [학생실습] 8월 일별 기온 데이터를 상자 그림으로 표현하기

```python
import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)
    
    
# 월별 데이터를 담을 리스트 생성
temp = []

for i in range(31):
    temp.append([])

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        if row[0].split("-")[1] == "08":
            day = row[0].split("-")[2]
            temp[int(day)-1].append(float(row[4]))  # temp[0].append() : 8월 1일 데이터

        
# 기본으로 그리니 잘 안 보여서        
plt.figure(figsize=(20,10), dpi=300)        

# showfliers : 이상치 값 안 보이게
plt.boxplot(temp, showfliers=False)
plt.show()
```
