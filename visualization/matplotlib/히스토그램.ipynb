```python
# 임의의 수를 뽑는 랜덤함수를 활용해 간단한 주사위 시뮬레이션 작성
# 임의로 추출된 주사위 수를 histogram 으로 작성하기



import random

dice = list()

for i in range(10):
    dice.append(random.randint(1,6))
    
print(dice)
```

```python
plt.hist(dice)  # 
plt.show()    # 하단의 시작 숫자는 그래프를 그릴 숫자를 참고하여 자동으로 맞춰줌
```

```python
# bins 옵션 : 가로축의 구간 개수 설정

plt.hist(dice, bins = 6) # bins default = 10
plt.show()
```

```python
# 주사위를 100번 던져서 나온 값 히스토그램으로 그리기

dice = list()

for i in range(100):
    dice.append(random.randint(1,6))

    
plt.hist(dice, bins = 6)
plt.show()
```

```python
# 기온 데이터를 히스토그램으로 표현하기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)


temp = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        temp.append(float(row[4]))

plt.hist(temp, bins=100, color="red")
plt.show()
```

```python
# 8월 데이터만 추출해서 히스토그램으로 그리기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)


temp = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        if row[0].split("-")[1] == "08":
            temp.append(float(row[4]))

plt.hist(temp, bins=100, color="red")
plt.show()
```

#### [학생 실습] 1월과 8월의 데이터를 히스토그램으로 시각화하기

```python
import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

plt.rc('font', family="D2Coding")

temp_08 = []
temp_01 = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        month = row[0].split("-")[1]
        if month == "01": 
            temp_01.append(float(row[4]))
        if month == "08":
            temp_08.append(float(row[4]))

plt.hist(temp_01, bins=100, color="red", label="1월")
plt.hist(temp_08, bins=100, color="blue", label="8월")
plt.legend()
plt.show()
```
