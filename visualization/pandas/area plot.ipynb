## area plot 산점도
- NaN 값이 있는 경우 0으로 자동 채워줌
- 0으로 채워주는 것이 싫다면 fillna(), dropna()를 통해서 전처리 하면 됨
- 자동으로 누적시켜서 그려줌

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제 해결
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.head()
```

```python
df.plot.area()
```

```python
# 기본으로 그려주는 누적 그래프 해제

df.plot.area(stacked=False)
```

```python
# 격자 무늬를 원한다면

df.plot.area(stacked=False, grid=True)
```
