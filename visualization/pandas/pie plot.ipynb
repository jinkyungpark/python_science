Seaborn 에서는 파이차트를 지원하지 않는다고? <a href='https://github.com/mwaskom/seaborn/issues/766'>Pie charts? · Issue #766 · mwaskom/seaborn</a>

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제 해결
mpl.rcParams['axes.unicode_minus'] = False
```

```python
series = pd.Series(3 * np.random.rand(4),
                      index=['a', 'b', 'c', 'd'], name='series')
series.plot.pie(figsize=(6, 6))   
```

```python
df = pd.DataFrame(3 * np.random.rand(4, 2),
                     index=['a', 'b', 'c', 'd'], columns=['x', 'y'])
  

df.plot.pie(subplots=True, figsize=(8, 4))
```

```python
# % 값 표시, 컬러 변경, font size 변경

series.plot.pie(labels=['AA', 'BB', 'CC', 'DD'], colors=['r', 'g', 'b', 'c'],
             autopct='%.2f', fontsize=20, figsize=(6, 6))
```

```python
# index를 줘서 특정 영역만 가지고 그리기

series = pd.Series([0.1] * 4, index=['a', 'b', 'c', 'd'], name='series2')

series.plot.pie(figsize=(6, 6))
```

```python
# dataFrame 을 이용한 차트 그리기

df = pd.DataFrame(
    {
        'mess': [0.330, 4.87, 5.97],
        'radius': [2439.7, 6051.8, 6378.1]
    },
    index=['mercury', 'Venus', 'Earth'])
plot = df.plot.pie(y='mess', figsize=(5, 5))
```

```python
plot = df.plot.pie(subplots=True, figsize=(6, 3))
```
