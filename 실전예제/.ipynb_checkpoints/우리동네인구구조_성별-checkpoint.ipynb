```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 분석할 데이터 로드

df = pd.read_csv("./resources/gender_202008.csv",
                 encoding="cp949", index_col=0)

df.head()
```

```python
df.shape
```

```python
dong = input("인구구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 ")

location = df.index.str.contains(dong)
df2 = df[location]
df2
```

```python
# 컬럼 삭제

del df2['2020년08월_남_총인구수'],df2['2020년08월_남_연령구간인구수'],df2['2020년08월_여_총인구수'],df2['2020년08월_여_연령구간인구수']
df2
```

```python
df2.columns = df2.columns.str.replace("2020년08월_","")
df2
```

```python
df_man = df2.iloc[:,0:101]
df_man.columns = df_man.columns.str.replace("남_","")
df_man
```

```python
df_woman = df2.iloc[:,101:]
df_woman.columns = df_woman.columns.str.replace("여_","")
df_woman
```

```python
df_man = df_man.transpose()
df_man
```

```python
df_man.columns = ["남성인구수"]
df_man
```

```python
df_man["남성인구수"]
```

```python
df_man = df_man['남성인구수'].astype("int64")
df_man
```

```python
plt.rc("font",family="Malgun Gothic")

df_man.plot.barh()
```

```python
df_woman = df_woman.transpose()
df_woman
```

```python
df_woman.columns = ["여성인구수"]
df_woman
```

```python
df_woman = df_woman['여성인구수'].astype("int64")
df_woman
```

```python
# - 값으로 변경

df_woman = -df_woman
df_woman
```

```python
plt.rc("font",family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False

df_woman.plot.barh(figsize=(30,10))
```

### 남/녀 같이 그래프 그리기


```python
df_gender = pd.concat([df_man, df_woman], axis=1)
df_gender
```

```python
df_gender.columns=["man","woman"]
df_gender
```

```python
df_gender.index
```

```python
# color 를 key,value 형태로 줄 수 있는 건 pandas 1.1.0 버전부터임(버전 업그레이드 반드시 해야 함)

plt.rc("font",family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False
df_gender.plot.barh(color={"man":"blue","woman":"red"},figsize=(30,20),title=dong+" 지역의 남녀 성별 인구 분포")
```
