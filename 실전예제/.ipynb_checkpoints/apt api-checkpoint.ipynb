### 규모 구분을 전용면적 컬럼으로 변경

규모구분 컬럼은 전용면적에 대한 내용임 - 컬럼명이 규모구분으로 되어 있어 직관적이지 않기 때문에
직관적인 컬럼을 만들어주고 전용면적, 이하, 초과 이런 단어들 정리

![nn](image/4.png)

```python
# 1. 규모구분에 어떤 값들이 고유 값인지 확인하기


df2['규모구분'].unique()
```

```python
# 2. 규모구분에 전용면적 이라는 글자를 빈 문자열로 변경
# pandas 에 있는 str.replace()를 사용하는 것

df2['전용면적'] = df2['규모구분'].str.replace("전용면적","")
df2['전용면적']
```

```python
# 3. 새로만든 전용면적 컬럼에서 이하는 공백, 초과는 물결로 변경하기

df2['전용면적'] = df2['전용면적'].str.replace("초과","~")
df2['전용면적'] = df2['전용면적'].str.replace("이하","")
df2['전용면적']
```

```python
# 4. 물결 뒤 공백 제거 및 앞 뒤 공백도 제거

df2["전용면적"] = (df2["전용면적"].str.replace(" ","").str.strip())
df2["전용면적"]
```

### 필요없는 컬럼 제거하기


```python
# 1. 전체 구조 확인하기

df2
```

```python
# 2. 필요없는 컬럼 제거하기
# axix - 0 : 행 / 1 : 열

df2 = df2.drop(
    [
        "규모구분",
        "분양가격(㎡)",
    ],
    axis=1,
)
df2
```

## 데이터 요약

### groupby로 데이터 집계하기

```python
# 1. 지역명으로 구분하고 평당분양가격 평균 구하기#

df2.groupby(["지역명"])["평당분양가격"].mean()
```

```python
# 2. 전용 면적으로 분양가격 평균 구하기

df2.groupby(["전용면적"])["평당분양가격"].mean()
```

```python
# 3. 지역명, 전용면적으로 구분하고 평당분양가격 평균 구하기

df2.groupby(["지역명","전용면적"])["평당분양가격"].mean()
```

```python
# 4. 전용면적,지역명으로 구분하고 평당분양가격 평균 + unstack()
# 뒤에 오는 컬럼을 열 명으로 사용 가능

df2.groupby(["전용면적","지역명"])["평당분양가격"].mean().unstack()
```

```python
# 4-1. 전용면적,지역명으로 구분하고 평당분양가격 평균 + unstack() + round()

df2.groupby(["전용면적","지역명"])["평당분양가격"].mean().unstack().round()
```

```python
# 5. 연도, 지역명으로 평당분양가격 평균 구하기

df2.groupby(["연도","지역명"])["평당분양가격"].mean()
```

```python
df2.groupby(["연도","지역명"])["평당분양가격"].mean().unstack()
```

```python
# T : transport

df2.groupby(["연도","지역명"])["평당분양가격"].mean().unstack().T
```

### pivot_table 사용하기

* 엑셀의 pivot 테이블과 유사

```python
# 1. 기본으로 평균값으로 지정되어 있으며 수치데이터를 모두 평균을 구해서 보여줌

pd.pivot_table(df2, index=['지역명'])
```

```python
# 2. 보고 싶은 컬럼만 보기

pd.pivot_table(df2, index=['지역명'], values=['평당분양가격'])
```

```python
# 3. 집계함수 지정 - mean, sum, count

pd.pivot_table(
    df2,
    index=[
        "지역명"
    ],
    values=[
        "평당분양가격"
    ],
    aggfunc="sum",
)
```

```python
# 전용 면적으로 분양가격 평균 구하기

pd.pivot_table(
    df2,
    index=[
        "전용면적"
    ],
    values=[
        "평당분양가격"
    ],
)
```

```python
# 3. 지역명, 전용면적으로 구분하고 평당분양가격 평균 구하기
# pd. 으로 시작해도 되고 df로 시작해도 됨

df2.pivot_table(
    index=[
        "전용면적",
        "지역명",
    ],
    values=[
        "평당분양가격"
    ],
)
```

```python
# 4. 전용면적,지역명으로 구분하고 평당분양가격 평균 + unstack()
# 뒤에 오는 컬럼을 열 명으로 사용 가능
# df2.groupby(["전용면적","지역명"])["평당분양가격"].mean().unstack()

df2.pivot_table(
    index=[
        "전용면적"
    ],
    values=[
        "평당분양가격"
    ],
    columns="지역명",
)
```

```python
# 4-1. 전용면적,지역명으로 구분하고 평당분양가격 평균 + unstack() + round()
# df2.groupby(["전용면적","지역명"])["평당분양가격"].mean().unstack().round()

df2.pivot_table(
    index=[
        "전용면적"
    ],
    values=[
        "평당분양가격"
    ],
    columns="지역명",
).round()
```

```python
# 5. 연도, 지역명으로 평당분양가격 평균 구하기
# df2.groupby(["연도","지역명"])["평당분양가격"].mean()

p = df2.pivot_table(
    index=[
        "연도",
        "지역명",
    ],
    values=[
        "평당분양가격"
    ],
)
p
```

```python
# 잘려진 데이터 들을 보고 싶다면

p.loc[2018]
```

## 연도와 월을 분리하기
* pandas의 string-handling 사용하기 : https://pandas.pydata.org/pandas-docs/stable/reference/series.html#string-handling

```python
date = "2013년12월"
date
```

```python
# split 을 통해 "년"을 기준으로 텍스트 분리
# df2는 나눠서 저장되어 있음

date2 = date.split("년")
```

```python
# 리스트의 인덱싱을 사용해서 연도만 가져오기
date2[0]
```

```python
# 리스트의 인덱싱을 사용해서 월 가져오고 "월" 글자는 제거
date2[-1].replace("월","")
```

```python
# 위의 개념을 이용해서 parse_year() 함수를 작성
# 연도만 반환하고 반환하는 데이터는 int 타입이 되도록 하기

def parse_year(date):
    return int(date.split("년")[0])

year = parse_year(date)
print(type(year))
year
```

```python
# 위의 개념을 이용해서 parse_month() 함수를 작성
# 월만 반환하고 반환하는 데이터는 int 타입이 되도록 하기

def parse_month(date):
    return int(date.split("년")[-1].replace("월",""))

month = parse_month(date)
print(type(month))
month
```

```python
# df_first_melt 변수에 담긴 데이터프레임에서 apply를 활용해 연도만 추출해서 새로운 컬럼에 담기

df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)

df_first_melt.head(1)
```

```python
# df_first_melt 변수에 담긴 데이터프레임에서 apply를 활용해 월만 추출해서 새로운 컬럼에 담기

df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
```

```python
# df2 와 병합을 하기 위해서는 컬럼의 이름이 같아야 하므로 확인

df2.sample()
```

## concat() 으로 두 개의 데이터 프레임 하나로 합치기

```python
# 최근 데이터가 담긴 df2에는 전용면적이 있음
# 이전 데이터에는 전용면적이 없기 때문에 "전체" 만 사용
# loc를 사용해서 전체에 해당하는 면적만 copy로 복사해서 df_last_prepare 변수에 담기

df2["전용면적"] == "전체"
```

```python
df2[df2["전용면적"] == "전체"]
```

```python
# 컬럼으로 사용할 컬럼명 리스트로 생성
cols = ["지역명","연도","월","평당분양가격"]
cols
```

```python
# df2 에서 원하는 컬럼만 추출하기
# copy()를 해줘야 df_last_prepare 에서 데이터가 수정이 되어도 df2 에 영향을 미치지 않음

df_last_prepare = df2.loc[df2["전용면적"] == "전체", cols].copy()
df_last_prepare
```

```python
# df_first_melt에서도 필요한 컬럼만 추출하기
df_first_melt.sample()
```

```python
# 필요한 컬럼만 추출한 후 df_first_prepare 에 담기

df_first_prepare = df_first_melt[cols].copy()
df_first_prepare
```

```python
# 컬럼명이 같다면 다른 작업을 크게 할 것이 없음

df_total = pd.concat([df_first_prepare,df_last_prepare])
df_total.head()
```

```python
# 제대로 합쳐졌는지 확인
df_total.shape
```

```python
df_total.tail()
```

```python
# 연도별로 데이터가 몇개씩 있는지  
# sort=False 없으면 최신 데이터부터 보여짐


df_total["연도"].value_counts(sort=False)
```

## pivot_table 사용하기

```python
# 연도를 인덱스로, 지역명을 컬럼으로 평당분양가격을 피봇 테이블로 그려보기
# pd.pivot : 연산은 하지 않고 데이터 변경
# pd.pivot_table : 연산도 하고 데이터도 변경

t = pd.pivot_table(df_total, index="연도", columns="지역명", values="평당분양가격").round()
t
```

```python
# 위의 표로는 데이터가 한눈에 들어오지 않음
# 위에서 그린 피봇 테이블을 히트맵으로 표현해 보기
sns.heatmap(t)
```

* 색상이 옅을 수록 높은 상태임
* 여러 색상이 섞여서 구별이 잘 안되기 때문에 하나의 색상 지정

```python
# 서울지역은 진하게 표시되고 경기도 약간 진하게 표시됨

sns.heatmap(t,cmap = "Blues")
```

```python
# 숫자 표시

plt.figure(figsize=(15,7))
sns.heatmap(t, cmap = "Blues", annot = True, fmt=".0f")
```

```python
# 바뀐 행과 열을 히트맵으로 표현하기
# transpose를 를 사용하여 행과 열 변경
# t.transpose() == t.T  같은 개념

plt.figure(figsize=(15,7))
sns.heatmap(t.T, cmap = "Blues", annot = True, fmt=".0f")
```

* 서울 지역의 경우 2013 ~ 2020 년으로 갈 수록 색상이 진해지는 것을 볼 수 있다 => 해가 갈수록 평당분양가가 올라간다.

```python
# groyp by로 그려보기, 인덱스에 ["연도","지역명"] 을 넣고 그리기

df_total.groupby(["연도","지역명"])["평당분양가격"].mean()
```
