#### csv 파일로 저장

```python
dict_data = {
    "c0": [
        1,
        2,
        3,
    ],
    "c1": [
        4,
        5,
        6,
    ],
}

df = pd.DataFrame(dict_data)
df
```

```python
# 데이터 프레임 csv 파일로 저장
# index를 False로 줘야 컬럼명에 , 들어가는 거 방지

df.to_csv(
    "./resources/df1.csv",
    index=False,
)
```

```python
# 데이터 프레임 excel 파일로 저장

df.to_excel("./resources/df1.xlsx")
```
