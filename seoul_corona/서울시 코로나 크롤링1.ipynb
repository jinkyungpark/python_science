# 서울시에서 공개한 코로나 19 발생 현황 분석

## 데이터 로드

```python
# 라이브러리 

import pandas as pd
```

```python
# 크롤링으로 가져올 url 주소 입력

url = "https://www.seoul.go.kr/coronaV/coronaStatus.do"
url
```

```python
# url을 통해서 들어가서 table 태그로 구성되어 있는 정보일 때 가능함
# read_html 은 사이트 내의 html 태그를 불러옴

pd.read_html?
```

```python
table = pd.read_html(url)
table
```

```python
# 읽어온 table이 몇 개인지 확인하기

len(table)
```

```python
# table 변수 안에 있는 값들을 확인하며 어떤 테이블을 선택할 것인지 알아보기
table[0]
```

```python
table[1]
```

```python
table[2]
```

```python
# table[2] 는 옆으로 기니까 세로로 해서 보기

table[2].transpose()        # table[2].T
```

```python
table[3]
```

```python
table[4]
```

```python
# 찾은 테이블을 데이터프레임에 넣어주기

df = table[3]
df.shape
```

```python
# 미리보기

df.head()
```

```python
# 마지막 확진일을 파일명에 써주기 위해 . 을 _로 변경
# 확장자와 구분이 쉽게 되도록 _로 변경

last_day = df.loc[0, "확진일"]
last_day
```

```python
last_day = last_day.replace(".","_")
last_day
```

```python
# 파일명을 만들어 주기 f"" (format string 의미)

file_name = f"seoul-covid19_{last_day}.csv"
file_name
```

```python
# csv 파일로 저장
df.to_csv(file_name)
```

```python
# 제대로 저장되었는지 다시 확인하기
# Unnamed : 기존에 있던 인덱스 값이 들어가서 나옴

pd.read_csv(file_name)
```

```python
# csv 저장시 인덱스는 저장하지 말라고 하기

df.to_csv(file_name, index=False)
```

```python
pd.read_csv(file_name)
```
