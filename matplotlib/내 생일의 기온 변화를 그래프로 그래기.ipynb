#### 데이터 추출하기

```python
# 기온만 추출하여 리스트에 저장하기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)


temp = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 기온 데이터만 리스트에 추가
for row in data:
    if row and row[4] != '':   # 주피터 노트북에서 실행할 때는  row 가 비어있지 않으면에 대한 코드가 꼭 필요함
        temp.append(float(row[4]))

print(temp)
```

```python
# 기온 데이터가 몇 개나 되는지 알아보기


len(temp)
```

#### 데이터 시각화하기

```python
import matplotlib.pyplot as plt
```

```python
# 리스트에 있는 온도를 차트로 나타내기

plt.plot(temp, "red")   # or plt.plot(temp, "r")
plt.show()
```

```python
# 각 년도의 8월 데이터만 추출하여 그래프 그리기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

temp_8 = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 8월 데이터 추출하기
for row in data:
    if row and row[4] != "":
        month = row[0].split("-")[1]
        if month == "08":
            temp_8.append(float(row[4]))

print(temp_8)
```

```python
plt.plot(temp_8, "hotpink")
plt.show()
```

```python
# 각 년도의 내 생일 온도를 그래프로 나타내기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

temp_8 = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)

# 8월 데이터 추출하기
for row in data:
    if row and row[4] != "":
        month = row[0].split("-")[1]
        day = row[0].split("-")[2]
        if month == "11" and day == "05":
            temp_8.append(float(row[4]))

print(temp_8)
```

```python
plt.plot(temp_8, "darkgreen")
plt.show()
```

```python
# 1990년 이후의 2/14 날짜의 기온 데이터를 추출해서 그래프 작성


import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

temp_8 = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)


high = []
low = list()

for row in data:
    if row and row[3] != "" and row[4] != "":       
        if int(row[0].split("-")[0]) >= 1990:
            if row[0].split("-")[1] == "02" and row[0].split("-")[2] == "14":
                # 최고 기온
                high.append(float(row[4]))
                # 최저기온
                low.append(float(row[3]))
            

plt.plot(high, "red", label="최고온도")
plt.plot(low, "blue", label="최저온도")
plt.legend()
plt.show()
```

#### [학생들 실습] 본인의 태어난 연도 이후 기온 데이터(최저,최고) 추출 후 그래프 작성

```python
# 본인의 태어난 연도 이후 생일날 데이터만 추출해서 최고기온과 최저 기온 그리기

import csv
f = open("../resources/ta_20200929111004.csv")
data = csv.reader(f)

temp_8 = []

# 상단의 쓸데없는 부분 제거하기
for i in range(8):
    next(data)


high = []
low = list()

for row in data:
    if row and row[3] != "" and row[4] != "":       
        if int(row[0].split("-")[0]) >= 1983:
            if row[0].split("-")[1] == "11" and row[0].split("-")[2] == "05":
                # 최고 기온
                high.append(float(row[4]))
                # 최저기온
                low.append(float(row[3]))
            
print(high)
print(low)
```

```python
# 위에서 추출한 데이터를 차트로 그려보기


plt.plot(high, "red", label="최고온도")
plt.plot(low, "blue", label="최저온도")
plt.legend()
plt.show()
```
