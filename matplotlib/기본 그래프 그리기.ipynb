#### plot 함수 기본 그래프 그리기
라이브러리 import <br>
plot([x축 데이터],[y축 데이터])<br>
show() : 그래프 보여주기

```python
# 선 그리기 - 1차원 리스트를 넘기는 경우 y 축에 할당됨

plt.plot([10, 20, 30, 40])
plt.show()
```

```python
# 선 그리기 - 두 개의 1차원 리스트를 인자로 주는 경우 x, y 축으로 그려짐

plt.plot([1,2,3,4],[12,43,25,15])
plt.plot()
```

#### plot 함수 기본 그래프 그리기 - 타이틀 입력

```python
plt.title("legend")
plt.plot([10, 20, 30, 40])
plt.show()
```

#### plot 함수 기본 그래프 그리기 - 범례

```python
plt.title("legend")
plt.plot([10, 20, 30, 40], label = "asc")
plt.plot([40, 30, 20, 10], label = "desc")
plt.legend()
plt.show()
```

```python
plt.title("legend")
plt.plot([10, 20, 30, 40], label = "asc")
plt.plot([40, 30, 20, 10], label = "desc")
# plt.legend(loc=5)
plt.legend(loc=1) 
plt.show()
```

#### plot 함수 기본 그래프 그리기 - 색상 변경

```python
plt.title("color")
plt.plot([10, 20, 30, 40], label = "asc", color="skyblue")
plt.plot([40, 30, 20, 10], "pink", label = "desc")  # 색상명만 주는 경우는 반드시 두번째 인자로
plt.legend() 
plt.show()
```

```python
# 그래프 직선 모양 변경하기 : linestyle or ls

plt.title("linestyle")
plt.plot([10, 20, 30, 40], label = "dashed", color="red", linestyle="--")
plt.plot([40, 30, 20, 10], "blue", label = "datted", ls=":")  # 색상명만 주는 경우는 반드시 두번째 인자로
plt.legend() 
plt.show()
```
